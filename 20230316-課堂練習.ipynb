1. import library

```python
import numpy as np
import matplotlib.pyplot as plt
```

2. 隨機生成訓練數據 X:0-100取隨機 y:2*X+1+隨機生成數

```python
#1.隨機生成訓練數據
np.random.seed(0)
n_samples = 100
X = np.random.randn(n_samples)# 隨機生成的訓練數據
y=2*X+1+np.random.randn(n_samples)#隨機生成的label數據
```

3.定義簡單回歸模型、loss function和梯度下降法

```python
def linear_regression(X, y, lr=0.1, n_iterations=10):#定義一個function名字叫linear_regression

    b0 = 0 #偏差(截距)初始值設為0
    w1 = 0 #權重(斜率)初始值設為0
    
    #宣告空的 list
    loss_history = [] #存放每一次iteration的loss值
    b0_history = [] #存放每一次iteration的偏差(截距)值
    w1_history = [] #存放每一次iteration的權重(斜率)值



  
    
    for i in range(n_iterations):
        
        y_pred = b0 +w1*X #定義簡單回歸模型
        loss = np.mean((y_pred-y)**2) #定義loss function (這裡採用MSE)
       
        #==========以下開始定義gradient descent梯度下降法的計算步驟================

        grad_b0=np.mean(2*(y_pred-y))# be斜率
        grad_w1 = np.mean(2*X*(y_pred-y))#w1斜率
        b0 = b0 - lr*grad_b0
        w1 = w1 - lr*grad_w1

        #==========以上結束定義gradient descent梯度下降法的計算步驟================

        loss_history.append(loss)
        b0_history.append(b0)
        w1_history.append(w1)
        
    return b0, w1, loss_history, b0_history, w1_history
```

4.上面隨機生成訓練數據帶入回歸模型，進行訓練

```python
b0,w1,loss_history,w0_history,w1_history = linear_regression(X,y)
```

5.畫loss fuction趨勢圖和回歸模型圖

```python
plt.figure(figsize=(10,5))
plt.plot(loss_history)
plt.xlabel('Iteration')
plt.ylabel('MSE Loss')
plt.title('Loss Trend')

plt.figure(figsize=(10,5))
plt.plot(w0_history)
plt.xlabel('Iteration')
plt.ylabel('wo')
plt.title('we Trend')

plt.plot(w1_history)
plt.xlabel('Iteration')
plt.ylabel('w1')
plt.title('w1 Trend')

plt.figure(figsize=(10,5))
plt.scatter(X,y)
plt.plot(X, b0 + w1*X, color='r')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression')

plt.show()
```
